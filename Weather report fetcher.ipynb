```python
import requests

def get_weather(city):
    # API key for OpenWeatherMap
    api_key = "68c097cd7c242636508b6acd21b0998b" 
    base_url = "http://api.openweathermap.org/data/2.5/weather?"
    
    # Constructing the API URL
    complete_url = f"{base_url}q={city}&appid={api_key}&units=metric"
    
    try:
        # Fetching data from the API
        response = requests.get(complete_url)
        data = response.json()
        
        # Checking if the city was found (cod 200 means success)
        if data["cod"] == 200:
            main = data["main"]
            weather = data["weather"][0]
            
            # Formatting and displaying the output
            print(f"\n--- Current Weather in {city.capitalize()} ---")
            print(f"Temperature: {main['temp']}°C")
            print(f"Feels Like: {main['feels_like']}°C")
            print(f"Humidity: {main['humidity']}%")
            print(f"Description: {weather['description'].capitalize()}")
        else:
            print("\nError: City not found. Please check the spelling.")
            
    except Exception as e:
        print(f"\nAn error occurred: {e}")

if __name__ == "__main__":
    city_name = input("Enter the city name to check weather: ")
    get_weather(city_name)
```
