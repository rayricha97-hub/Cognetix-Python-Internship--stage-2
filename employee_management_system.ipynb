```python
# import sqlite3

# 1. Database connection aur Table creation
def connect_db():
    conn = sqlite3.connect('company.db')
    cursor = conn.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS employees 
                      (id INTEGER PRIMARY KEY, name TEXT, email TEXT, dept TEXT)''')
    conn.commit()
    conn.close()

# 2. CRUD Operations
def add_employee(name, email, dept):
    conn = sqlite3.connect('company.db')
    cursor = conn.cursor()
    cursor.execute("INSERT INTO employees (name, email, dept) VALUES (?, ?, ?)", (name, email, dept))
    conn.commit()
    conn.close()
    print("Employee added successfully!")

def view_employees():
    conn = sqlite3.connect('company.db')
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM employees")
    rows = cursor.fetchall()
    print("\n--- Employee Records ---")
    for row in rows:
        print(f"ID: {row[0]} | Name: {row[1]} | Email: {row[2]} | Dept: {row[3]}")
    conn.close()

# 3. Main Interface
def main():
    connect_db()
    while True:
        print("\n--- Employee Management System ---")
        print("1. Add Employee\n2. View All Employees\n3. Exit")
        choice = input("Enter choice (1-3): ")
        
        if choice == '1':
            name = input("Enter Name: ")
            email = input("Enter Email: ")
            dept = input("Enter Dept: ")
            add_employee(name, email, dept)
        elif choice == '2':
            view_employees()
        elif choice == '3':
            break
        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()
```
